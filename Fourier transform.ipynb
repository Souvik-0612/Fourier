```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import sin, cos, pi, sqrt, exp
plt.style.use(['science', 'notebook', 'grid'])
```

## Fourier Transform of Continuous Function

>### In case of  even function :
$$\mathcal{G(\alpha)} = \sqrt{\frac{2}{\pi}} \int_{0}^{\infty}\mathcal{F(x)}\cos(\alpha x)dx$$
$$\mathcal{F(x)} = \sqrt{\frac{2}{\pi}} \int_{0}^{\infty}g(\alpha)\cos(\alpha x)d\alpha$$

>### In case of odd function :
$$\mathcal{G(\alpha)} = \sqrt{\frac{2}{\pi}} \int_{0}^{\infty}\mathcal{F(x)}\sin(\alpha x)dx$$
$$\mathcal{F(x)} = \sqrt{\frac{2}{\pi}} \int_{0}^{\infty}\mathcal{G(\alpha)}\sin(\alpha x)d\alpha$$

```python
def Trap(xi, xf, h, f):
    xx, yy = [], []
    while (abs(xi - xf) >= h):
        xx.append(xi)
        y = f(xi)
        yy.append(y)
        xi += h
        
    sum1 = sum(yy[i] for i in range(1, len(yy)-2))
    
    return 0.5*h*(yy[0]+ 2*sum1 +yy[len(yy)-1])
```

## Even function fourier transform

> ### Fourier transform of: 
$$\mathcal{F(x)} = e^{-x^2}$$

```python
ai = -6
af = 6
da = 0.01

# Empty lists:
aa = []; gg = []; ff = []

#F(x)
def f(x):
    return np.exp(-x**2)
    
#Fourier Transform    
def g(a): return Trap(-5, 5, da, lambda x: f(x)*cos(a*x))/sqrt(2*pi)  #As it is a even function

while abs(ai - af) >= da:
    ff.append(f(ai))
    gg.append(g(ai))
    aa.append(ai)
    ai += da
plt.figure(figsize=(14, 6))
plt.plot(aa, ff, "--", label = r"$\mathcal{F(x)}$")
plt.plot(aa, gg, label = r"$\mathcal{G(\alpha)}$")
plt.legend(loc = 'best')
plt.show()
```

> ### Fourier transform of: 
$$ \mathcal{F(x)}=  \left\{
\begin{array}{ll}
      |x| & [-3, 3] \\
      0 & [else] \\
\end{array} 
\right. 
$$

```python
ai = -6
af = 6
da = 0.01

# Empty lists:
aa = []; gg = []; ff = []

#F(x)
def f(x):
    if abs(x)< 3: return abs(x)
    else: return 0
    
#Fourier Transform    
def g(a): return Trap(-3, 3, da, lambda x: f(x)*cos(a*x))/sqrt(2*pi)  #As it is a even function

while abs(ai - af) >= da:
    ff.append(f(ai))
    gg.append(g(ai))
    aa.append(ai)
    ai += da
plt.figure(figsize=(14, 6))
plt.plot(aa, ff, "--", label = r"$\mathcal{F(x)}$")
plt.plot(aa, gg, label = r"$\mathcal{G(\alpha)}$")
plt.legend(loc = 'best')
plt.show()
```

## Odd functions

> ### Fourier transform of: 
$$ \mathcal{F(x)}=  \left\{
\begin{array}{ll}
      -1 & [-3, 0] \\
      1 & [0, 3] \\
      0 & [else] \\
\end{array} 
\right. 
$$

> ### Fourier transform of: 
$$ \mathcal{F(x)}=  \left\{
\begin{array}{ll}
      1 & [-0.4, 0.4] \\
      0 & [else] \\
\end{array} 
\right. 
$$

```python
ai = -6
af = 6
da = 0.01

# Empty lists:
aa = []; gg = []; ff = []

#F(x)
def f(x):
    if abs(x)<0.4: return 1
    else: return 0
    
#Fourier Transform    
def g(a): return Trap(-3, 3, da, lambda x: f(x)*cos(a*x))/sqrt(2*pi)  #As it is a even function

while abs(ai - af) >= da:
    ff.append(f(ai))
    gg.append(g(ai))
    aa.append(ai)
    ai += da
plt.figure(figsize=(14, 6))
plt.plot(aa, ff, "--", label = r"$\mathcal{F(x)}$")
plt.plot(aa, gg, label = r"$\mathcal{G(\alpha)}$")
plt.legend(loc = 'best')
plt.show()
```

```python
ai = -6
af = 6
da = 0.01

# Empty lists:
aa = []; gg = []; ff = []

#F(x)
def f(x):
    if -3 < x < 0: return -1
    elif 0 < x < 3: return 1
    else: return 0
    
#Fourier Transform    
def g(a): return Trap(-3, 3, da, lambda x: f(x)*sin(a*x))/sqrt(2*pi)  #As it is a odd function

while abs(ai - af) >= da:
    ff.append(f(ai))
    gg.append(g(ai))
    aa.append(ai)
    ai += da
plt.figure(figsize=(14, 6))
plt.plot(aa, ff, "--", label = r"$\mathcal{F(x)}$")
plt.plot(aa, gg, label = r"$\mathcal{G(\alpha)}$")
plt.legend(loc = 'best')
plt.show()
```

## Remark: 
 
1. As I minimize the thickness of the $\mathcal{F(x)}$ the fourier transform $\mathcal{G(x)}$ spread more!
2. We can relate this visual to Uncertainty principle i.e, if we know the position exactly we cannot measure the momentum. So from this what we can refer is that *The Fourier tansform of position is momentum and vice versa!*.

# Even-odd funtion

> ### Fourier transform of: 
$$\mathcal{F(x)} = \cos(f_1x)+\sin(f_2 x)$$
$f_1$  and $f_2$ being the frequencies of the given signal

```python
ai = 0
af = 5
da = 0.01

# Empty lists:
aa = []; ggR = []; ggI = []; ff = []

#F(x)
def F(x):
    return cos(3*x)+sin(4*x)  #Note:here f1 = 3; f2 = 4
    


#Fourier Transform    
def gR(a): return Trap(-20, 20, da, lambda x: F(x)*cos(a*x))/sqrt(2*pi)  #As it is a even function
def gI(a): return Trap(-20, 20, da, lambda x: F(x)*sin(a*x))/sqrt(2*pi)  #As it is a odd function

while abs(ai - af) >= da:
    ff.append(F(ai))
    ggI.append(gI(ai))
    ggR.append(gR(ai))
    aa.append(ai)
    ai += da
    
    
plt.figure(figsize=(14, 6))
plt.plot(aa, ff, "--", label = r"$\mathcal{F(x)}$")
plt.plot(aa, ggR, label = r"$Re(\mathcal{G(\alpha)})$")
plt.plot(aa, ggI, label = r"$Im(\mathcal{G(\alpha)})$")
plt.legend(loc = 'best')
plt.show()
```

## Remarks:
I intentionally write the $\mathcal{F(x)}$ as $\cos(3x)+\sin(4 x)$ to see that **The fourier tansfrom gives the frequencies**.And as expected Fourier transform gives the appropiate *frequencies*.
